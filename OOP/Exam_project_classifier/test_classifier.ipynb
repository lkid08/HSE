```python
from transformers import pipeline


class Classifier:
    def __init__(self, model_name):
        self.classifier = pipeline('zero-shot-classification', model=model_name)
        

    def classify_text(self, text):
        candidate_labels = ["политика", "экономика", "развлечения", "экология", "технологии"]

        return self.classifier(text, candidate_labels, multi_label=False)
```

```python
classifier = Classifier('MoritzLaurer/mDeBERTa-v3-base-xnli-multilingual-nli-2mil7')
```
