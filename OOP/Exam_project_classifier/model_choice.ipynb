```python
from huggingface_hub import HfApi, ModelFilter
api = HfApi()
```

```python
models = api.list_models(
    filter=ModelFilter(
		task="text-classification",
		language='ru'
	)
)
```

```python
list_of_models = [model for model in models]
list_of_models[:10]
```

```python
len(list_of_models)
```
